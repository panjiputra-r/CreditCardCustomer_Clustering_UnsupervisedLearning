```python
import pickle
import pandas as pd
import numpy as np
```

```python
# open Model
with open("model.pkl", "rb") as f1:
    km = pickle.load(f1)


# open scaller
with open("scaler.pkl", "rb") as f2:
    scaler = pickle.load(f2)

    
# open PCA
with open("pca.pkl", "rb") as f3:
    pca = pickle.load(f3)

```

```python
np.random.seed(40)


data = {
    'BALANCE': np.random.uniform(0, 7563.976453, 10),
    'BALANCE_FREQUENCY': np.random.uniform(0, 1, 10),
    'PURCHASES': np.random.uniform(0, 1890.85, 10),
    'ONEOFF_PURCHASES': np.random.uniform(0, 3000.50, 10),
    'INSTALLMENTS_PURCHASES': np.random.uniform(0, 1000.00, 10),
    'CASH_ADVANCE': np.random.uniform(0, 24267.415750, 10),
    'PURCHASES_FREQUENCY': np.random.uniform(0, 1, 10),
    'ONEOFF_PURCHASES_FREQUENCY': np.random.uniform(0, 1, 10),
    'PURCHASES_INSTALLMENTS_FREQUENCY': np.random.uniform(0, 1, 10),
    'CASH_ADVANCE_FREQUENCY': np.random.uniform(0, 0.54, 10),
    'CASH_ADVANCE_TRX': np.random.uniform(0, 80, 10),
    'PURCHASES_TRX': np.random.uniform(0, 50, 10),
    'CREDIT_LIMIT': np.random.uniform(100, 12100.0, 10),
    'PAYMENTS': np.random.uniform(0, 2923.967334, 10),
    'MINIMUM_PAYMENTS': np.random.uniform(56.720210, 1892.941133, 10),
    'PRC_FULL_PAYMENT': np.random.uniform(0, 0.90, 10),

    }

dataInf = pd.DataFrame(data)
dataInf
```

```python
# Scaling
data_scaled = scaler.transform(dataInf)
```

```python
# PCA
data_scaled_pca = pca.transform(data_scaled)
```

```python
# melakukan predict
pred = km.predict(data_scaled_pca)
pred
```

```python
# memasukkan hasil predict ke dataframe
dataInf['cluster'] = pred

dataInf
```
